```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
csv_data = pd.read_excel("AirQualityUCI.xlsx")
```

```python
csv_data
```

```python
csv_data.describe()
```

```python
csv_data = csv_data[csv_data>0].dropna()#缺失值处理
```

```python
csv_data.describe()
```

```python
train_data = csv_data[['CO(GT)']]
```

```python
train_target = csv_data[['PT08.S1(CO)']]
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(train_data, train_target, test_size=0.3)
```

```python
LR = LinearRegression().fit(x1_train, y1_train)
```

```python
s1 = LR.score(x1_train, y1_train)
s2 = LR.score(x1_test,y1_test)
```

```python
result = LR.predict(x1_test)
```
